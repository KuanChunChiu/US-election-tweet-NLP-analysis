### Experiments and preliminary results

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
democrat_20 = pd.read_csv("Data/cleaned/democrat_2020_cleaned.csv")
republican_20 = pd.read_csv("Data/cleaned/republican_2020_cleaned.csv")
democrat_24 = pd.read_csv("Data/cleaned/democrat_2024_cleaned.csv")
republican_24 = pd.read_csv("Data/cleaned/republican_2024_cleaned.csv")

demo_20_shape = democrat_20.shape
repu_20_shape = republican_20.shape
demo_24_shape = democrat_24.shape
repu_24_shape = republican_24.shape

print ("Dimension of the 4 datasets:\n")
print (f"Democrat 2020: {demo_20_shape}")
print (f"Republican 2020: {repu_20_shape}")
print (f"Democrat 2024: {demo_24_shape}")
print (f"Republican 2024: {repu_24_shape}")
```

```python
for df in [democrat_20, republican_20, democrat_24, republican_24]:
    df["Party"] = df["label"].str.title()

tweets_20 = pd.concat([democrat_20, republican_20], axis=0)
tweets_24 = pd.concat([democrat_24, republican_24], axis=0)

shape_20 = tweets_20.shape
shape_24 = tweets_24.shape

print ("Dimension of the 2 merged datasets:\n")
print (f"2020 tweets: {shape_20}")
print (f"2024 tweets: {shape_24}")
```

```python
demo_perc_20 = round((demo_20_shape[0]*100) / shape_20[0], 1)
repu_perc_20 = round((repu_20_shape[0]*100) / shape_20[0], 1)
demo_perc_24 = round((demo_24_shape[0]*100) / shape_24[0], 1)
repu_perc_24 = round((repu_24_shape[0]*100) / shape_24[0], 1)

print ("Percentage of democrat and republican tweets in each dataset:\n")
print (f"2020 tweets: {demo_perc_20}% democrat, {repu_perc_20}% republican")
print (f"2024 tweets: {demo_perc_24}% democrat, {repu_perc_24}% republican")
```

```python
demo_20_tweets = list(tweets_20[tweets_20["Party"] == "Democrat"]["text"])
repu_20_tweets = list(tweets_20[tweets_20["Party"] == "Republican"]["text"])
demo_24_tweets = list(tweets_24[tweets_24["Party"] == "Democrat"]["text"])
repu_24_tweets = list(tweets_24[tweets_24["Party"] == "Republican"]["text"])

demo_20_tweets_wrd_cnt = [len(str(tweet).strip().split(" ")) for tweet in demo_20_tweets]
repu_20_tweets_wrd_cnt = [len(str(tweet).strip().split(" ")) for tweet in repu_20_tweets]
demo_24_tweets_wrd_cnt = [len(str(tweet).strip().split(" ")) for tweet in demo_24_tweets]
repu_24_tweets_wrd_cnt = [len(str(tweet).strip().split(" ")) for tweet in repu_24_tweets]

demo_20_avg_tweet_wrd_cnt = round(np.mean(demo_20_tweets_wrd_cnt).item(), 1)
repu_20_avg_tweet_wrd_cnt = round(np.mean(repu_20_tweets_wrd_cnt).item(), 1)
demo_24_avg_tweet_wrd_cnt = round(np.mean(demo_24_tweets_wrd_cnt).item(), 1)
repu_24_avg_tweet_wrd_cnt = round(np.mean(repu_24_tweets_wrd_cnt).item(), 1)

print (f"Average word count of each democrat tweet in the 2020 dataset: {demo_20_avg_tweet_wrd_cnt}")
print (f"Average word count of each republican tweet in the 2020 dataset: {repu_20_avg_tweet_wrd_cnt}")
print (f"Average word count of each democrat tweet in the 2024 dataset: {demo_24_avg_tweet_wrd_cnt}")
print (f"Average word count of each republican tweet in the 2024 dataset: {repu_24_avg_tweet_wrd_cnt}")
```
